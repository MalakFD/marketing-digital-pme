```python
import pandas as pd

# Charger le dataset (mettre le chemin exact indiqué par Kaggle)
df = pd.read_csv('/kaggle/input/marketing/Marketing.csv')

# Afficher les 5 premières lignes
df.head()
```

```python
# Vérifier les valeurs manquantes
print("Valeurs manquantes par colonne :")
print(df.isnull().sum())

# Supprimer les lignes avec valeurs manquantes
df = df.dropna()

# Vérifier les types de colonnes
print("Types de colonnes :")
print(df.dtypes)
```

```python
# Statistiques descriptives
df.describe()
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x='mark_spent', y='revenue', data=df)
plt.title("Impact de l'investissement marketing sur le revenu")
plt.xlabel("Dépenses marketing")
plt.ylabel("Revenu")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['revenue'], bins=10, edgecolor='black')
plt.title("Distribution du revenu des campagnes marketing")
plt.xlabel("Revenu")
plt.ylabel("Nombre de campagnes")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.hist(df['revenue'], bins=10, edgecolor='black')
plt.title("Distribution du revenu des campagnes marketing")
plt.xlabel("Revenu")
plt.ylabel("Nombre de campagnes")
plt.show()
```

```python
# Création de catégories selon les dépenses marketing
df['marketing_level'] = pd.qcut(
    df['mark_spent'],
    3,
    labels=['Faible', 'Moyen', 'Élevé']
)

# Vérifier le résultat
df[['mark_spent', 'marketing_level']].head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x='marketing_level', y='revenue', data=df)
plt.title("Revenu selon le niveau d'investissement marketing")
plt.xlabel("Niveau d'investissement marketing")
plt.ylabel("Revenu")
plt.show()
```

# Analyse de l’impact du marketing digital sur la performance

## Objectif
Analyser la relation entre les dépenses marketing digitales et le revenu généré
par les campagnes des PME marocaines.

## Méthodologie
- Nettoyage des données
- Analyse exploratoire
- Visualisation des relations clés
- Modélisation par régression linéaire
- Segmentation des campagnes marketing

## Résultats
Les visualisations et la régression montrent qu’une augmentation des dépenses
marketing est associée à une augmentation du revenu, confirmant l’impact positif
du marketing digital sur la performance.

## Conclusion
L’investissement en marketing digital joue un rôle important dans l’amélioration
de la performance des PME.
